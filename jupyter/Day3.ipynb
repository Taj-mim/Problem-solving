```python
#arg and kwargs

def avg(*num):   #arg
    sum=0
    for n in num :
        sum=sum+n
    return sum / len(num)
print(avg(1,2,3,4,5,6))

def avgg(**num): #kwargs
    sum=0
    for keys in num :
        sum=sum+num[keys]
    return sum / len(num)
print(avgg(a=10,b=20,c=80))
```

```python

def info(names,*num,**number):
    f,l=name(names)
    print("my first name is {} .and my last name {}".format(f,l))
    average=avg(*num)
    print("my average number is {}".format(average))
    ag=avgg(**number)
    print("kwargs average is {}".format(ag))

info("fatema taj mim",10,20,30,40,50,am=10,bm=20,cm=30)
```

```python
#recursive function


def factorial(n):
    if n==1:
        return 1
    elif n==0:
        return 1
    elif n<0:
        print("cannot compute")
    else:
        return (n*(factorial(n-1)))

print(factorial(5))


def fibonacci_recursive(n):
    if n <= 0:
        return "Invalid input"
    elif n == 1:
        return 0
    elif n == 2:
        return 1
    else:
        return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)

print(fibonacci_recursive(10))  
```
